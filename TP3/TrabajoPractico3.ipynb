```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Inciso A: Distribución Uniforme (Min: 0, Max: 1)
valores_uniformes = np.random.uniform(0, 1, 1000)
plt.subplot(221)  # Subplot en la posición superior izquierda
plt.hist(valores_uniformes, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución Uniforme')

# Inciso B: Distribución Normal (Media: 0, Desvío: 1)
normal_values = np.random.normal(0, 1, 1000)
plt.subplot(222)  # Subplot en la posición superior derecha
plt.hist(normal_values, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribución Normal')

# Inciso C: Distribución de Poisson (λ = 6)
poisson_values = np.random.poisson(6, 1000)
plt.subplot(223)  # Subplot en la posición inferior izquierda
plt.hist(poisson_values, bins=30, color='salmon', edgecolor='black')
plt.title('Distribución Poisson')

# Inciso D: Distribución Exponencial (λ = 3/4)
exponential_values = np.random.exponential(scale=34, size=1000)
plt.subplot(224)  # Subplot en la posición inferior derecha
plt.hist(exponential_values, bins=30, color='gold', edgecolor='black')
plt.title('Distribución Exponencial')

# Ajustar espaciado entre subplots
plt.tight_layout()

# Mostrar la figura con los histogramas
plt.show()
```

```python
def transformar_a_exponencial(datos,beta=12):
    return -np.log(1 - datos) / beta

valores_transformados = transformar_a_exponencial(valores_uniformes)
```

```python
# Inciso 3 b)

# Gráfico para observar la distribución exponencial resultante
plt.figure(figsize=(8, 6))
plt.hist(valores_transformados, bins=30, color='orange', edgecolor='black')
plt.title('Distribución Exponencial')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Inciso 3 c) 

distribucion_exponencial = np.random.exponential(3, 100)
#Inciso 3 d)
distribucion_exponencial2 = np.random.exponential(3, 1000)
# Inciso 3 e) 

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(distribucion_exponencial, bins=30, color='purple', alpha=0.7)
plt.title('Distribución Exponencial con λ = 3 (n = 100)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(distribucion_exponencial2, bins=30, color='cyan', alpha=0.5)
plt.title('Distribución Exponencial con λ = 3 (n = 1000)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

<h3>Inciso 3 f)</h3>

- Media y dispersión: Dado que ambos conjuntos de datos tienen el mismo parámetro lambda (λ = 3) pero diferentes tamaños de muestra (n = 100 y n = 1000),  el histograma con n = 1000 tiene una distribución más cercana a la distribución teórica de la exponencial. Esto se debe  a que  a medida que el tamaño de la muestra aumenta, la distribución muestral se aproxima más a la distribución teorica de la población.

- Forma de la distribución: Dado que estamos trabajando con una distribución exponencial los valores se agrupen cerca de cero y disminuyan exponencialmente a medida que aumentan los valores. Sin embargo, debido a la aleatoriedad en la generación de datos  se puede observar que en la muestra con tamaño  de la muenstra n 100 se evidencia mas  cierta variabilidad en la forma exacta del histograma contrario  alo que ocurre con la el tamaño de muestra n=1000. 


```python

#Calcular la media muestral
media_muestral_3c = np.mean(distribucion_exponencial)
media_muestral_3d = np.mean(distribucion_exponencial2)

# Calcular el desvío estándar
desvio_estandar_3c = np.std(distribucion_exponencial)
desvio_estandar_3d = np.std(distribucion_exponencial2)

# Calcular la varianza muestral
varianza_muestral_3c = np.var(distribucion_exponencial, ddof=1)
varianza_muestral_3d = np.var(distribucion_exponencial2, ddof=1)

print("Resultados para distribucion_exponencial (n = 100):")
print("a) Media muestral:", media_muestral_3c)
print("b) Desvío estándar:", desvio_estandar_3c)
print("c) Varianza muestral:", varianza_muestral_3c)

print("\nResultados para distribucion_exponencial2 (n = 1000):")
print("a) Media muestral:", media_muestral_3d)
print("b) Desvío estándar:", desvio_estandar_3d)
print("c) Varianza muestral:", varianza_muestral_3d)
```
