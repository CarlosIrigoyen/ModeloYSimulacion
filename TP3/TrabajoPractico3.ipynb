```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar el tamao de la figura
plt.figure(figsize=(12, 8))

# Inciso A: Distribucin Uniforme (Min: 0, Max: 1)
valores_uniformes = np.random.uniform(0, 1, 1000)
plt.subplot(221)  # Subplot en la posicin superior izquierda
plt.hist(valores_uniformes, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribucin Uniforme')

# Inciso B: Distribucin Normal (Media: 0, Desvo: 1)
normal_values = np.random.normal(0, 1, 1000)
plt.subplot(222)  # Subplot en la posicin superior derecha
plt.hist(normal_values, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribucin Normal')

# Inciso C: Distribucin de Poisson ( = 6)
poisson_values = np.random.poisson(6, 1000)
plt.subplot(223)  # Subplot en la posicin inferior izquierda
plt.hist(poisson_values, bins=30, color='salmon', edgecolor='black')
plt.title('Distribucin Poisson')

# Inciso D: Distribucin Exponencial ( = 3/4)
exponential_values = np.random.exponential(scale=34, size=1000)
plt.subplot(224)  # Subplot en la posicin inferior derecha
plt.hist(exponential_values, bins=30, color='gold', edgecolor='black')
plt.title('Distribucin Exponencial')

# Ajustar espaciado entre subplots
plt.tight_layout()

# Mostrar la figura con los histogramas
plt.show()
```

```python
def transformar_a_exponencial(datos,beta=12):
    return -np.log(1 - datos) / beta

valores_transformados = transformar_a_exponencial(valores_uniformes)
```

```python
# Inciso 3 b)

# Grfico para observar la distribucin exponencial resultante
plt.figure(figsize=(8, 6))
plt.hist(valores_transformados, bins=30, color='orange', edgecolor='black')
plt.title('Distribucin Exponencial')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Inciso 3 c) 

distribucion_exponencial = np.random.exponential(3, 100)
#Inciso 3 d)
distribucion_exponencial2 = np.random.exponential(3, 1000)
# Inciso 3 e) 

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(distribucion_exponencial, bins=30, color='purple', alpha=0.7)
plt.title('Distribucin Exponencial con  = 3 (n = 100)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(distribucion_exponencial2, bins=30, color='cyan', alpha=0.5)
plt.title('Distribucin Exponencial con  = 3 (n = 1000)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

<h3>Inciso 3 f)</h3>

- Media y dispersin: Dado que ambos conjuntos de datos tienen el mismo parmetro lambda ( = 3) pero diferentes tamaos de muestra (n = 100 y n = 1000),  el histograma con n = 1000 tiene una distribucin ms cercana a la distribucin terica de la exponencial. Esto se debe  a que  a medida que el tamao de la muestra aumenta, la distribucin muestral se aproxima ms a la distribucin teorica de la poblacin.

- Forma de la distribucin: Dado que estamos trabajando con una distribucin exponencial los valores se agrupen cerca de cero y disminuyan exponencialmente a medida que aumentan los valores. Sin embargo, debido a la aleatoriedad en la generacin de datos  se puede observar que en la muestra con tamao  de la muenstra n 100 se evidencia mas  cierta variabilidad en la forma exacta del histograma contrario  alo que ocurre con la el tamao de muestra n=1000. 


```python

#Calcular la media muestral
media_muestral_3c = np.mean(distribucion_exponencial)
media_muestral_3d = np.mean(distribucion_exponencial2)

# Calcular el desvo estndar
desvio_estandar_3c = np.std(distribucion_exponencial)
desvio_estandar_3d = np.std(distribucion_exponencial2)

# Calcular la varianza muestral
varianza_muestral_3c = np.var(distribucion_exponencial, ddof=1)
varianza_muestral_3d = np.var(distribucion_exponencial2, ddof=1)

print("Resultados para distribucion_exponencial (n = 100):")
print("a) Media muestral:", media_muestral_3c)
print("b) Desvo estndar:", desvio_estandar_3c)
print("c) Varianza muestral:", varianza_muestral_3c)

print("\nResultados para distribucion_exponencial2 (n = 1000):")
print("a) Media muestral:", media_muestral_3d)
print("b) Desvo estndar:", desvio_estandar_3d)
print("c) Varianza muestral:", varianza_muestral_3d)
```
