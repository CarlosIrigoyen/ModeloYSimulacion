```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Inciso A: Distribución Uniforme (Min: 0, Max: 1)
uniform_values = np.random.uniform(0, 1, 1000)
plt.subplot(221)  # Subplot en la posición superior izquierda
plt.hist(uniform_values, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución Uniforme')

# Inciso B: Distribución Normal (Media: 0, Desvío: 1)
normal_values = np.random.normal(0, 1, 1000)
plt.subplot(222)  # Subplot en la posición superior derecha
plt.hist(normal_values, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribución Normal')

# Inciso C: Distribución de Poisson (λ = 6)
poisson_values = np.random.poisson(6, 1000)
plt.subplot(223)  # Subplot en la posición inferior izquierda
plt.hist(poisson_values, bins=30, color='salmon', edgecolor='black')
plt.title('Distribución Poisson')

# Inciso D: Distribución Exponencial (λ = 1/34)
exponential_values = np.random.exponential(scale=34, size=1000)
plt.subplot(224)  # Subplot en la posición inferior derecha
plt.hist(exponential_values, bins=30, color='gold', edgecolor='black')
plt.title('Distribución Exponencial')

# Ajustar espaciado entre subplots
plt.tight_layout()

# Mostrar la figura con los histogramas
plt.show()
```
