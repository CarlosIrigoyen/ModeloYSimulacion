import numpy as np
import matplotlib.pyplot as plt
EXP=30
COR=100

# Hotel

```python

# Inciso 1: El tiempo promedio de finalización del proyecto y el IC. Con el 99% de probabilidad (2,57).
def confidence_interval(n_sample, mean, std):
    z_value = 2.57
    coeff = 1.0
    z_term = coeff * (z_value * std / (n_sample ** 0.5))
    inf = mean - z_term
    sup = mean + z_term
    return inf, sup, z_term

# Inciso 2: Calcule el porcentaje de criticidad tienen los diferentes accesos.
def porcentajeCritico(tiempo, tiempo_total):
    return tiempo * 100 / tiempo_total

tiempos_ejecucion_exp = []
for exp in range(EXP):
    tiempos_ejecucion_corr = []
    criticidades = {'sup': [], 'medio': [], 'inf': []}
    for cor in range(COR):
        # Generar números aleatorios para cada tarea
        a = np.random.uniform(2, 4)
        b = np.random.uniform(3, 6)
        c = np.random.uniform(2, 5)
        d = np.random.uniform(3, 6)
        e = np.random.uniform(4, 8)
        f = np.random.uniform(3, 7)
        
        tiempo_total = a + b + c + d + e + f  # En minutos

        # Tiempo superior
        tiempo_sup = a + b + c
        criticidades['sup'].append(porcentajeCritico(tiempo_sup, tiempo_total))

        # Tiempo Medio
        tiempo_medio = a + d + e + f 
        criticidades['medio'].append(porcentajeCritico(tiempo_medio, tiempo_total))

        # Tiempo Inferior
        tiempo_inf = f
        criticidades['inf'].append(porcentajeCritico(tiempo_inf, tiempo_total))
        
        tiempos_ejecucion_corr.append(tiempo_total)

    tiempos_ejecucion_exp.append(np.mean(tiempos_ejecucion_corr))

# Inciso 1: Mostrar variables
media = np.mean(tiempos_ejecucion_exp)
desvio = np.std(tiempos_ejecucion_exp)
ic_inf, ic_sup, z_term = confidence_interval(COR, media, desvio)

print("# Inciso 1: El tiempo promedio de finalización del proyecto y el IC.")
print("El tiempo promedio de finalización del proyecto es:", media)
print("Desvio:", desvio)
print("Z-Term:", z_term)
print("Intervalo de confianza (99%): [{:.2f}, {:.2f}]".format(ic_inf, ic_sup))
print()

# Inciso 2: Mostrar variables
for key in criticidades.keys():
    print("# Inciso 2: Porcentaje de criticidad para", key)
    print(criticidades[key])
    print()

# Inciso 3: Graficar histogramas
print("# Inciso 3: Histogramas")
# Histograma de la distribución del tiempo de realización del proyecto (corridas)
plt.hist(tiempos_ejecucion_corr, bins=20, color="red")
plt.title("Tiempo total de corridas")
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de la distribución del tiempo de realización del proyecto (experimentos)
plt.hist(tiempos_ejecucion_exp, bins=20, color="green")
plt.title("Tiempo total de experimentos")
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.show()




```
