```python
import numpy as np
import statistics as stats
import seaborn as sns
import matplotlib.pyplot as plt
CANT_EXPERIMENTOS = 30
CANT_CORRIDAS = 100
duraciones = []
promedios_experimentos = []
criticidad_as = 0
criticidad_am = 0
criticidad_ai = 0

```

# Hotel

```python

#Obtiene, suma y retorna las demoras de las tareas a,b,c del acceso superior
def calcular_acceso_superior():
    tarea_a = round(np.random.uniform(2, 4), 2)
    tarea_b = round(np.random.uniform(3, 6), 2)
    tarea_c = round(np.random.uniform(2, 5), 2)
    return tarea_a + tarea_b + tarea_c

#Obtiene, suma y retorna las demoras de las tareas d y e del acceso medio
def calcular_acceso_medio():
    tarea_d = np.random.uniform(3,6)
    tarea_e = np.random.uniform(2,5)
    return tarea_d + tarea_e

#Obteniene, suma y retorna las demoras de las tareas f y g del acceso inferior
def calcular_acceso_inferior():
    tarea_f = round(np.random.uniform(4, 8), 2)
    tarea_g = round(np.random.uniform(3, 7), 2)
    return tarea_f + tarea_g

#Obtiene y retorna las demoras de los accesos
def preparar_desayuno():
    acceso_superior = calcular_acceso_superior()
    acceso_medio = calcular_acceso_medio()
    acceso_inferior = calcular_acceso_inferior()
    return [acceso_superior, acceso_medio, acceso_inferior]    

#Calcula el intervalo de confianza
def calcular_IC(n, media, desvio, z):
    l = z * (desvio / np.sqrt(n))
    return media - l ,  media + l 

def graficar_histograma(a):
    sns.set()
    sns.histplot(a, color="violet")
    plt.show()

if __name__ == "__main__":
    # Simulacion de los 30 experimentos de 100 corridas
    for i in range(CANT_EXPERIMENTOS):
        duracion_experimento = 0
        for j in range(CANT_CORRIDAS):
            accesos = preparar_desayuno()
            duracion_corrida = sum(accesos)
            duraciones.append(duracion_corrida)
            duracion_experimento += duracion_corrida

            #Obtengo el acceso más crítico y aumento su contador
            criticidad = max(accesos)        
            if criticidad == accesos[0]:
                criticidad_as += 1
            if criticidad == accesos[1]:
                criticidad_am += 1
            if criticidad == accesos[2]:
                criticidad_ai += 1

        promedios_experimentos.append(duracion_experimento / CANT_CORRIDAS)

    media = stats.mean(duraciones)
    media = round(media, 2)
    z = 2.57     
    limite_inferior, limite_superior = calcular_IC(len(duraciones), media, stats.stdev(duraciones), z)
    limite_inferior = round(limite_inferior, 2)
    limite_superior = round(limite_superior, 2)

    #Calculo los porcentajes de la criticidad de cada acceso
    porcentaje_criticidad_as = criticidad_as * 100 / (CANT_EXPERIMENTOS * CANT_CORRIDAS) 
    porcentaje_criticidad_am = criticidad_am * 100 / (CANT_EXPERIMENTOS * CANT_CORRIDAS)
    porcentaje_criticidad_ai = criticidad_ai * 100 / (CANT_EXPERIMENTOS * CANT_CORRIDAS)

    # Normalizar los porcentajes para que sumen 100%
    total_criticidad = porcentaje_criticidad_as + porcentaje_criticidad_am + porcentaje_criticidad_ai
    porcentaje_criticidad_as = round((porcentaje_criticidad_as / total_criticidad) * 100, 2)
    porcentaje_criticidad_am = round((porcentaje_criticidad_am / total_criticidad) * 100, 2)
    porcentaje_criticidad_ai = round((porcentaje_criticidad_ai / total_criticidad) * 100, 2)
    total=porcentaje_criticidad_as+porcentaje_criticidad_am+porcentaje_criticidad_ai
    print("El tiempo promedio de finalización del proyecto es de ", media, " minutos")
    print("El Intervalo de confianza tiene un límite inferior = ", limite_inferior, " y un límite superior = ", limite_superior)
    print("El porcentaje de criticidad del acceso superior es de ", porcentaje_criticidad_as, "%")
    print("El porcentaje de criticidad del acceso medio es de ", porcentaje_criticidad_am, "%")
    print("El porcentaje de criticidad del acceso inferior es de ", porcentaje_criticidad_ai, "%")
    print("Porcentaje total:", total )
    graficar_histograma(duraciones)
    graficar_histograma(promedios_experimentos)
```

# Empresa Constructora

```python

import numpy as np
import statistics as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Función para construir una casa
def construir_casa():
    # Definición de las tareas con sus respectivos rangos de tiempo
    tarea_a = np.random.uniform(2, 4)
    tarea_b = np.random.uniform(3, 5)
    tarea_c = np.random.uniform(1, 2)
    tarea_d = np.random.uniform(4, 8)
    tarea_e = np.random.uniform(3, 6)
    tarea_f = np.random.uniform(2, 5)
    tarea_g = np.random.uniform(2, 4)
    tarea_h = np.random.uniform(1, 3)
    tarea_i = np.random.uniform(2, 4)
    tarea_j = np.random.uniform(2, 3)
    
    # Acceso superior: suma de tareas a, b, c, d, e, i, f, j del acceso medio más tarea g
    acceso_superior = tarea_a + tarea_b + tarea_c + tarea_d + tarea_e + tarea_i + tarea_f + tarea_j + tarea_g
    
    # Acceso medio: suma de tareas a, b, c, d, e, f, i, j
    acceso_medio = tarea_a + tarea_b + tarea_c + tarea_d + tarea_e + tarea_f + tarea_i + tarea_j
    
    # Acceso inferior: suma de tareas a, b, c, d, e, f, i, j del acceso medio más tarea h
    acceso_inferior = tarea_a + tarea_b + tarea_c + tarea_d + tarea_e + tarea_f + tarea_i + tarea_j + tarea_h
    
    return [acceso_superior, acceso_medio, acceso_inferior]

# Función para calcular el intervalo de confianza
def calcular_IC(n, media, desvio, z):
    l = z * (desvio / np.sqrt(n))
    return media - l ,  media + l 

# Función para graficar un histograma
def graficar_histograma(data, xlabel):
    sns.set()
    sns.histplot(data, color="violet")
    plt.xlabel(xlabel)
    plt.ylabel("Frecuencia")
    plt.title("Histograma de " + xlabel)
    plt.show()

if __name__ == "__main__":
    duraciones = []
    promedios_experimentos = []
    criticidad_as = 0
    criticidad_am = 0
    criticidad_ai = 0

    # Simulación de los experimentos y corridas previamente definidos
    for i in range(CANT_EXPERIMENTOS):
        duracion_experimento = 0
        for j in range(CANT_CORRIDAS):
            accesos = construir_casa()
            duracion_corrida = sum(accesos)
            duraciones.append(duracion_corrida)
            duracion_experimento += duracion_corrida

            # Obtengo el acceso más crítico y aumento su contador
            criticidad = max(accesos)        
            if criticidad == accesos[0]:
                criticidad_as += 1
            if criticidad == accesos[1]:
                criticidad_am += 1
            if criticidad == accesos[2]:
                criticidad_ai += 1

        promedios_experimentos.append(duracion_experimento / CANT_CORRIDAS)

    # Cálculo de estadísticas
    media = stats.mean(duraciones)
    z = 2.57
    limite_inferior, limite_superior = calcular_IC(len(duraciones), media, stats.stdev(duraciones), z)     
    
    # Cálculo de porcentajes de criticidad de cada acceso
    total_corridas = CANT_EXPERIMENTOS * CANT_CORRIDAS
    porcentaje_criticidad_as = criticidad_as * 100 / total_corridas 
    porcentaje_criticidad_am = criticidad_am * 100 / total_corridas
    porcentaje_criticidad_ai = criticidad_ai * 100 / total_corridas
    
    # Impresión de resultados
    print("El tiempo promedio de finalización del proyecto es de", media, "días")
    print("El Intervalo de confianza tiene un límite inferior =", limite_inferior, "y un límite superior =", limite_superior)
    print("El porcentaje de criticidad del acceso superior es de", porcentaje_criticidad_as, "%")
    print("El porcentaje de criticidad del acceso medio es de", porcentaje_criticidad_am, "%")
    print("El porcentaje de criticidad del acceso inferior es de", porcentaje_criticidad_ai, "%")

    # Visualización de resultados
    graficar_histograma(duraciones, "Tiempo de finalización del proyecto (días)")
    graficar_histograma(promedios_experimentos, "Tiempo promedio de finalización del proyecto por experimento (días)")

```
